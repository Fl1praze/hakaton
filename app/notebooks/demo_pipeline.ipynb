## 1Ô∏è‚É£ –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –≥–æ—Ç–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω


```python
import sys
sys.path.append('..')

from processor import extract_text_from_pdf
from hybrid_bert_model import HybridBERTExtractor
import json
from pathlib import Path

print("‚úÖ –ò–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–ª–∏ extract_text_from_pdf –∏ HybridBERTExtractor")
```

---
## 2Ô∏è‚É£ –ó–∞–≥—Ä—É–∂–∞–µ–º ML‚Äë–º–æ–¥–µ–ª—å


```python
print("üîÑ –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è HybridBERTExtractor...")
extractor = HybridBERTExtractor()
info = extractor.get_info()
print("‚úÖ –ú–æ–¥–µ–ª—å –∑–∞–≥—Ä—É–∂–µ–Ω–∞:")
print(info)
```

---
## 3Ô∏è‚É£ –í—ã–±–∏—Ä–∞–µ–º –ª—é–±–æ–π —á–µ–∫ –∏–∑ `app/pdf_for_study/`


```python
pdf_dir = Path("../pdf_for_study")
pdf_files = list(pdf_dir.glob("*.pdf"))

print(f"–ù–∞–π–¥–µ–Ω–æ PDF —Ñ–∞–π–ª–æ–≤: {len(pdf_files)}")

# –±–µ—Ä–µ–º –ø–µ—Ä–≤—ã–π —Ñ–∞–π–ª –¥–ª—è –ø—Ä–∏–º–µ—Ä–∞
test_pdf = pdf_files[0]
print(f"–ë—É–¥–µ–º –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å: {test_pdf.name}")
```

---
## 4Ô∏è‚É£ –ò–∑–≤–ª–µ–∫–∞–µ–º —Ç–µ–∫—Å—Ç


```python
with open(test_pdf, 'rb') as f:
    pdf_bytes = f.read()

text = extract_text_from_pdf(pdf_bytes, verbose=False)

print(f"–î–ª–∏–Ω–∞ —Ç–µ–∫—Å—Ç–∞: {len(text)} —Å–∏–º–≤–æ–ª–æ–≤")
print("\n–ü–µ—Ä–≤—ã–µ 500 —Å–∏–º–≤–æ–ª–æ–≤:\n")
print(text[:500])
```

---
## 5Ô∏è‚É£ –ü—Ä–∏–º–µ–Ω—è–µ–º HybridBERTExtractor (ML + regex)


```python
result = extractor.predict(text)

print("–ù–∞–π–¥–µ–Ω–Ω—ã–µ –æ—Å–Ω–æ–≤–Ω—ã–µ –ø–æ–ª—è:")
for key in ['inn', 'total', 'date', 'time', 'vendor', 'email', 'phone']:
    if key in result and result[key]:
        print(f"  {key}: {result[key]}")

if 'ml_confidence' in result:
    print("\nML confidence (—É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏):")
    for k, v in result['ml_confidence'].items():
        print(f"  {k}: {v:.2f} ({v*100:.0f}%)")
```

---
## 6Ô∏è‚É£ JSON, –∫–æ—Ç–æ—Ä—ã–π –±–æ—Ç –ø–æ–ª—É—á–∞–µ—Ç –æ—Ç API


```python
json_response = json.dumps(result, ensure_ascii=False, indent=2)
print(json_response)
```
